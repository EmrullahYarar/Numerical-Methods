```python
def f1(x):
    f = (x**2)-(5*x)+3
    return f
```

```python
print('f1(0)=', f1(0), 'f1(1)=', f1(1))
```

```python
f1(8)
```

```python
import matplotlib.pyplot as plt
x = []
y = []
for i in range (-100,100):
    x.append(i)
    y.append(f1(i))
plt.plot(x,y)
```

```python
def der_f1(x):
    der = 2*x-5
    return der
print(der_f1(2))
```

```python
print('der_f1(0)=', der_f1(0),',', 'der_f1(1)=', der_f1(1),',', 'der_f1(10)=', der_f1(10),',', 'der_f1(100)=', der_f1(100))
```

```python
x = []
y = []
z = []
for i in range (-100,100):
    x.append(i)
    y.append(f1(i))
    z.append(der_f1(i))
plt.title("f1(x) and its derivative")
plt.plot(x,y, "r")
plt.plot(x,z)
```

```python
def num_der_f1(x, h =0.000000001):
    der = (f1(x+h)-f1(x))/h
    return der
```

```python
num_der_f1(5)
```

```python
print('num_der_f1(0)=',num_der_f1(0), 'num_der_f1(1)=',num_der_f1(1), 'num_der_f1(0)=',num_der_f1(0), 'num_der_f1(0)=',num_der_f1(0))
```

```python
x = []
y = []
z = []
for i in range (-10,10):
    x.append(i)
    y.append(der_f1(i))
    z.append(num_der_f1(i, h=1))
plt.plot(x,y, 'r')
plt.plot(x,z)
```

```python
import math
def f2(x):
    g= math.exp(x)-math.cos(x)
    return g
```

```python
f2(0), f2(1), f2(10)
```

```python
def der_f2(x):
    e= math.exp(x)+math.sin(x)
    return e
```

```python
der_f2(0), der_f2(1), der_f2(10) 
```

```python
def num_der_f2(x, h=0.0000001):
    z= (f2(x+h)-f2(x))/h
    return z
```

```python
num_der_f2(0), num_der_f2(1), num_der_f2(10)
```

```python
x = []
y = []
z = []
for i in range (-10,10):
    x.append(i)
    y.append(der_f1(i))
    z.append(num_der_f2(i, h=1))
plt.plot(x,y, 'r')
plt.plot(x,z)
```
