```python
import numpy as np
import pylab
```

```python
def func(x):
    return -0.1*x**4-0.15*x**3-0.5*x**2-0.25*x+1.2
func(5)
```

```python
def fdfunc(x):
    return -0.4*x**3-0.45*x**2-x-0.25
fdfunc(5)
```

```python
def sdfunc(x):
    return -1.2*x**2-0.9*x-1
sdfunc(5)
```

```python
def taylorh(f, d1_f, d2_f, x, n):
    h=x
    e=[]
    fapp=[]
    for i in range(n):
        f_a = f(x-h) + d1_f(x-h)*h + (d2_f(x-h)*h**2)/2
        fapp.append(f_a)
        e.append(abs(f(x)-f_a))
        h=h/2
    return fapp, e
```

```python
taylorh(func, fdfunc, sdfunc, 5, 3)
```

```python
xval, yval = taylorh(func, fdfunc, sdfunc, 5.0, 10)
xr = [1.0/(2**x) for x in range (10)]

pylab.plot(xr, yval, 'kx')
pylab.plot(xr, yval, 'm')
pylab.ylabel('absolute error')
pylab.xlabel('xr')
pylab.title('Plot of absolute error as h decreases')
pylab.semilogy()
```

```python
def f(x):
    return x**0.5
```

```python
def fdf(x):
    return 1/2*(x**1/2)
```

```python
def sdf(x):
    return 1/4*(x**3/2)    
```

```python
def tdf(x):
    return 3/8*(x**5/2)
```

```python
def taylorh(f, d1_f, d2_f, d3_f, x, n):
    h=x
    e=[]
    fapp=[]
    for i in range(n):
        f_a = f(x-h) + d1_f(x-h)*h + (d2_f(x-h)*h**2)/2 + (d3_f(x-h)*h**3)/6
        fapp.append(f_a)
        e.append(abs(f(x)-f_a))
        h=h/2
    return fapp, e
```

```python
xval, yval = taylorh(f, fdf, sdf, tdf, 5.0, 10)
xr = [1.0/(2**x) for x in range (10)]

pylab.plot(xr, yval, 'kx')
pylab.plot(xr, yval, 'm')
pylab.ylabel('absolute error')
pylab.xlabel('xr')
pylab.title('Plot of absolute error as h decreases')
pylab.semilogy()
```
